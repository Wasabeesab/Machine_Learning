```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

import Library yang akan digunakan 

```python
dataset = pd.read_csv('bank_customer.csv', sep = ';')
dataset
```

membaca dataset

```python
dataset.info()
```

informasi dataset

```python
categorical_features = [feature for feature in dataset.columns
                        if dataset[feature].dtypes == "O"]
dataset[categorical_features]
```

Kode pertama mengidentifikasi fitur-fitur yang memiliki tipe data kategorikal (misalnya, 'job', 'marital', 'education', dll.) menggunakan dataset.columns dan dataset[feature].dtypes == "O".

```python
for i in categorical_features:
    print(dataset[i].unique())
```

identifikasi data unique categorical 

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

for feature in categorical_features:
    print(dataset[feature].unique())
    dataset[feature] = label_encoder.fit_transform(dataset[feature])
    print(dataset[feature].unique())
```

Mengubah data kategorikal menjadi numerik

```python
dataset
```

menampilkan hasil pergantian data 

```python
x = dataset.drop(['y', 'age'], axis = 1)
y = dataset['y']
```

Kolom 'y' (target) dan 'age' dihapus dari dataset untuk memisahkan fitur (X) dan target (y). Hal ini dilakukan untuk mempersiapkan data agar bisa digunakan dalam pelatihan model

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.25,random_state = 42) 
```

Data dibagi menjadi set pelatihan dan set pengujian menggunakan train_test_split dengan proporsi 75% untuk pelatihan dan 25% untuk pengujian. Pembagian ini penting untuk memastikan bahwa model dapat diuji dengan data yang belum pernah dilihat sebelumnya.

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train, y_train)
```

Model yang digunakan dalam kode ini adalah GaussianNB dari sklearn.naive_bayes. Model ini bekerja dengan prinsip Naive Bayes, yang merupakan metode probabilistik untuk klasifikasi.

```python
y_pred_train = model.predict(x_train)
y_pred_test = model.predict(x_test)
```

Pengujian model dilakukan dengan menggunakan data pengujian (x_test) yang telah dibagi sebelumnya. Model melakukan prediksi pada data ini dengan menggunakan model.predict(x_test). Hasil prediksi kemudian dievaluasi menggunakan matriks kebingungan dan laporan klasifikasi untuk menilai performa model pada data yang tidak terlihat.

```python
model.n_features_in_
```

```python
x_baru = np.array([[35, 1, 1, 2, 0, 1200, 1, 0, 0, 5, 8, 300, 2, 30, 3]])
y_pred_baru = model.predict(x_baru)
y_pred_baru
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print('evaluasi training')
print(confusion_matrix(y_train, y_pred_train))
print(classification_report(y_train, y_pred_train))

print('evaluasi testing')
print(confusion_matrix(y_test, y_pred_test))
print(classification_report(y_test, y_pred_test))
```

Analisis akurasi model dilakukan dengan menggunakan Confussion Matrix dan clasification report
